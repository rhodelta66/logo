#Stappen na een Succesvolle respons

Als er een succesvolle respons is, dan heeft u een inboundverzoek gepost in Tracy WMSx. Er staat daar dan een te verwerken Bill of Lading inclusief alle bijbehorende details, zoals de naam van de Boot, de verwachte datum van aankomst, de Container(s) en hun inhoud.

##Inbound Transport
Op het moment dat het bekend is dat de container op transport wordt gezet van de haven terminal naar de Warehouselocatie, dan wordt er een bericht geplaatst in het douane tabel (wms_io). Hierdoor kan de douane afdeling dan alvast aan de slag, bijvoorbeeld door het aanvragen van een "toestemming tot openen" van de gesealde container. Want we weten nu dat de container snel aan de poort bij het warehouse zal staan.

##Aankomst van de container bij het warehouse
Als de container aankomt, dan wordt deze geopend en de inhoud wordt geinspecteerd op volledigheid. Eventuele mix-pallets worden omgezet naar enkelvoudige pallets. Dit betekent dat er op 1 pallet dan alleen identieke artikelen staan Met zelfde artikelcode / SKU.

Na controle van de pallets krijgen ze een sticker. Deze wordt ook wel de pallet-sticker genoemd. De pallet-sticker zorgt er voor dat de pallet getraceerd kan blijven worden gedurende de tijd dat deze in het warehouse staat.

Bij het printen van de stickers wordt ook bepaald op welke (loods-)positie in het warehouse de pallets worden opgeslagen. Als men de QR op de pallet-sticker scant, dan is de gekoppelde positie af te lezen. Uiteraard is het de bedoeling dat de pallets vervolgens op die positie worden neergezet.
